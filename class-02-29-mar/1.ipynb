```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

sentences = [
    "I am good.",
    "I am not bad."
]

# Convert sentences to bag-of-words vectors
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(sentences)

# Show vocabulary
print("Vocabulary:", vectorizer.get_feature_names_out())

# Show vectors
print("\nSentence vectors:")
print(X.toarray())

# Cosine similarity
sim = cosine_similarity(X[0], X[1])[0][0]
print("\nCosine similarity (CountVectorizer):", sim)
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

sentences = [
    "I am good.",
    "I am not bad."
]

# Load embedding model
model = SentenceTransformer("all-MiniLM-L6-v2")

# Create embeddings
embeddings = model.encode(sentences)

# Cosine similarity
sim = cosine_similarity([embeddings[0]], [embeddings[1]])[0][0]
print("Cosine similarity (Sentence Embeddings):", sim)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer

sentences = [
    "I am good.",
    "I am not bad."
]

# -----------------------------
# Method 1: CountVectorizer
# -----------------------------
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(sentences)

print("=== CountVectorizer ===")
print("Vocabulary:", vectorizer.get_feature_names_out())
print("Vectors:")
print(X.toarray())

count_sim = cosine_similarity(X[0], X[1])[0][0]
print("Cosine similarity:", count_sim)

# -----------------------------
# Method 2: Sentence Embeddings
# -----------------------------
model = SentenceTransformer("all-MiniLM-L6-v2")
embeddings = model.encode(sentences)

embed_sim = cosine_similarity([embeddings[0]], [embeddings[1]])[0][0]

print("\n=== Sentence Embeddings ===")
print("Embedding shape:", embeddings.shape)
print("Cosine similarity:", embed_sim)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

sentences = [
    "I am good.",
    "I am not bad."
]

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(sentences)

print("Vocabulary:", vectorizer.get_feature_names_out())
print("TF-IDF matrix:")
print(X.toarray())

sim = cosine_similarity(X[0], X[1])[0][0]
print("Cosine similarity:", sim)
```
